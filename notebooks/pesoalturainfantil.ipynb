### Importação das bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Carregamento dos dados
Dados de altura por idade - WHO percentiles boys Birth to 5 years:

https://www.who.int/childgrowth/standards/cht_lhfa_boys_p_0_5.pdf?ua=1

Dados de peso por idade - WHO percentiles boys Birth to 5 years: 

https://www.who.int/childgrowth/standards/cht_wfa_boys_p_0_5.pdf?ua=1

```python
dflhfa = pd.read_csv('../data/lhfa_boys_p_exp.csv', sep=';') # Dados de altura por idadde - OMS
dfwfa = pd.read_csv('../data/wfa_boys_p_exp.csv', sep=';') # Dados de peso por idadde - OMS
dfhenrique = pd.read_csv('../data/dadoshenrique.csv', sep=';') # Dados de Henrique
```

## Dados de Altura

### Carregando eixo X (dias)

```python
xlhfa = dflhfa['Day']
xlhHenrique = dfhenrique['Dias']
```

### Carregando eixo y (altura)

Representação dos percentis: P3, P15, P50, P85 e P97

```python
ylhfa3 = dflhfa['P3']
ylhfa15 = dflhfa['P15']
ylhfa50 = dflhfa['P50']
ylhfa85 = dflhfa['P85']
ylhfa97 = dflhfa['P97']
ylhHenrique = dfhenrique['Altura']
```

### Plotando gráfico

```python
plt.figure( figsize=(20, 12))

plt.plot(xlhfa, ylhfa3, color='red')
plt.plot(xlhfa, ylhfa15, color='orange')
plt.plot(xlhfa, ylhfa50, color='green')
plt.plot(xlhfa, ylhfa85, color='orange')
plt.plot(xlhfa, ylhfa97, color='red')

plt.xticks(np.arange(0, 1856, step=60))
plt.yticks(np.arange(0, 130, step=5))


plt.scatter(xlhHenrique, ylhHenrique, color='blue')

plt.xlabel('Dias')
plt.ylabel('Altura')
plt.title('Percentil - Altura (LH) - WHO')
plt.grid(True)
plt.show()
```

## Dados de Peso

### Carregando eixo X (dias)

```python
xwfa = dfwfa['Age']
xwHenrique = dfhenrique['Dias']
```

### Carregando eixo y (peso)

Representação dos percentis: P3, P15, P50, P85 e P97

```python
ywfa3 = dfwfa['P3']
ywfa15 = dfwfa['P15']
ywfa50 = dfwfa['P50']
ywfa85 = dfwfa['P85']
ywfa97 = dfwfa['P97']
ywHenrique = dfhenrique['Peso']
```

### Plotando gráfico

```python
plt.figure( figsize=(20, 12))

plt.plot(xwfa, ywfa3, color='red')
plt.plot(xwfa, ywfa15, color='orange')
plt.plot(xwfa, ywfa50, color='green')
plt.plot(xwfa, ywfa85, color='orange')
plt.plot(xwfa, ywfa97, color='red')

plt.xticks(np.arange(0, 1856, step=60))
plt.yticks(np.arange(0, 25, step=1))

plt.scatter(xwHenrique, ywHenrique, color='blue')

plt.xlabel('Dias')
plt.ylabel('Peso')
plt.title('Percentil - Peso (W) - WHO')
plt.grid(True)
plt.show()
```
